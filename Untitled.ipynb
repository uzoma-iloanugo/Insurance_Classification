```python
import os
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
from fancyimpute import IterativeImputer
import pickle
```

```python
raw_df = pd.read_csv('exercise_40_train.csv')
raw_df.head()
```

```python
# Separate Categorical Variables by selecting all rows ':', and filtering all columns where dtype is object
cat_df = raw_df.loc[:, raw_df.dtypes == object]
cat_df.head()
```

```python
num_df = raw_df.loc[:, raw_df.dtypes == np.float64]
num_df.head()
```

```python
# Describe categorical variable
cat_df.describe(include =[object])
```

```python
cat_df['x7'] = cat_df['x7'].str.replace('%', '').astype(float)
cat_df.head()
```

```python
cat_df['x19'] = cat_df['x19'].str.replace('\$', '').astype(float)
cat_df.head()
```

```python
# drop var x33 because it is constant
cat_df.drop(['x33'], axis = 1, inplace = True)
```

```python
cat_df.head()
```

```python
# add converted categorical var to num df
num_df['x7'] = cat_df['x7']
num_df['x19'] = cat_df['x19']
cat_df.drop(['x7', 'x19'], axis = 1, inplace = True)
```

```python

for col in cat_df:
    print(cat_df[col].value_counts(dropna=False), '\n')
```

```python
# Variable x3 has wednesdays and wed meaning the same. So, I have to replace and match them using mask operator

day_mask1 = cat_df['x3'] == 'Wednesday'
day_mask1.value_counts()
```

```python
cat_df['x3'][day_mask1] = 'Wed'
cat_df['x3'].value_counts()
```

```python
day_mask2 = cat_df['x3'] == 'Monday'
day_mask3 = cat_df['x3'] == 'Tuesday'
day_mask4 = cat_df['x3'] == 'Thursday'
day_mask5 = cat_df['x3'] == 'Friday'
day_mask6 = cat_df['x3'] == 'Saturday'
day_mask7 = cat_df['x3'] == 'Sunday'
```

```python
cat_df['x3'][day_mask2] = 'Mon'
cat_df['x3'][day_mask3] = 'Tue'
cat_df['x3'][day_mask4] = 'Thur'
cat_df['x3'][day_mask5] = 'Fri'
cat_df['x3'][day_mask6] = 'Sat'
cat_df['x3'][day_mask7] = 'Sun'
```

```python
cat_df['x3'].value_counts()
```

```python
# Perform inductive, single imputation with numerical df
# NoTE: I Will not attempt multiple imputation
# VERY IMPORTANT PROCESSSSSS!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

imp = IterativeImputer()
imp_num_df = pd.DataFrame(imp.fit_transform(num_df))
imp_num_df.index = num_df.index
imp_num_df.columns = num_df.columns
```

```python
imp_num_df
```

```python
# Save imputation fit

with open('imp', 'wb') as i:
    pickle.dump(imp, i, pickle.HIGHEST_PROTOCOL)
```

```python
# Use the most frequent occuring categorical variable for missing values

for col in cat_df:
    max_freq = cat_df[col].value_counts().index[0]
    cat_df[col][pd.isna(cat_df[col])] = max_freq
```

```python
cat_df.isna().sum()
```

```python
le_dict = {}
label_df = pd.DataFrame()

for col in cat_df:
    le = LabelEncoder()
    label_df[col] = le.fit_transform(cat_df[col])
    le_dict[col] = le
```

```python
# save the label encoder

with open('le_dict', 'wb') as l:
    pickle.dump(le_dict, l, pickle.HIGHEST_PROTOCOL)
```

```python
# Create dataframe of binary column representing instances of each category across multiple columns
# cat_df dimension 40000 by 9 - after encoding 40000 by 39!

enc = OneHotEncoder()
onehot_df = pd.DataFrame(enc.fit_transform(label_df).toarray())
```

```python
feat_names = enc.get_feature_names()
onehot_df.columns = feat_names
```

```python
# save OneHotEncoder var

with open('enc', 'wb') as e:
    pickle.dump(enc, e, pickle.HIGHEST_PROTOCOL)
```

```python
cat_df.shape
```

```python
# Scale the numeric variable

scale = StandardScaler()
scale_num_df = pd.DataFrame(scale.fit_transform(imp_num_df))
```

```python
# Pickle Scaler object

with open('scale', 'wb') as s:
    pickle.dump(scale, s, pickle.HIGHEST_PROTOCOL)
```

```python
x_df = pd.concat([scale_num_df, onehot_df], axis = 1)
```

```python
# Picke x_df

with open('x_df', 'wb') as x:
    pickle.dump(x_df, x, pickle.HIGHEST_PROTOCOL)
```

```python
# separate the target variable

y_df = raw_df['y'].astype('int')
```

```python
# save y var

with open('y_df', 'wb') as y:
    pickle.dump(y_df, y, pickle.HIGHEST_PROTOCOL)
```

```python


from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.decomposition import PCA
from sklearn.svm import SVC

pipe = Pipeline([('logit', LogisticRegression(C=10, solver='liblinear'))])

params = {}
params['logit__C'] = [0.01, 0.05, 0.1, 0.5, 1]
#params['pca__n_components'] = [15, 20, 25, 30, 35, 50]

grid1 = GridSearchCV(pipe, params, cv=5, scoring = 'roc_auc')
grid1.fit(x_df, y_df);
print(grid1.best_params_)
print(grid1.best_score_)
```

```python
from sklearn.svm import SVC

pipe = Pipeline([('svc', SVC())])

params = {}
params['svc__kernel'] = ['rbf']
params['svc__gamma'] = [1e-3, 1e-4]
params['svc__C'] = [1, 10, 100, 1000]

grid1 = GridSearchCV(pipe, params, cv=5, scoring = 'roc_auc')
grid1.fit(x_df, y_df);
print(grid1.best_params_)
print(grid1.best_score_)
```
